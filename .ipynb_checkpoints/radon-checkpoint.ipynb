```python
import tensorflow as tf
from tensorflow.keras.layers import Input
from tensorflow.keras.optimizers import Adam
# import keras.backend as K
from tensorflow.keras import backend as K
from tensorflow.keras.callbacks import Callback
from skimage.transform import radon, rescale
import matplotlib.pyplot as plt
import numpy as np
import cv2
from model import get_Model
import time
from rrmse import rrmse
t0 = time.time()

tf.random.set_seed(100)
print(tf.__version__)
```

```python
from scipy.io import loadmat
mat = loadmat('data/radon.mat')
print(mat)
A = mat['A']
print(type(A))
Sshape = mat['Sshape']

A = A.toarray()
A = np.asarray(A, order='C')
```

```python
gr_img = plt.imread('data/ChestPhantom.png')[:,:]
N, _ = gr_img.shape
```

```python
X = gr_img.reshape((N**2, 1))
S = A.dot(X)
radon = S.reshape((Sshape)[0])
print(radon.shape)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(8, 4.5))

ax1.set_title("Original")
ax1.imshow(gr_img, cmap=plt.cm.Greys_r)
ax2.set_title("Radon transform\n(Sinogram)")
ax2.set_xlabel("Projection angle (deg)")
ax2.set_ylabel("Projection position (pixels)")
ax2.imshow(radon, cmap=plt.cm.Greys_r)
#            extent=(0, 180, 0, S.shape[0]), aspect='auto')

fig.tight_layout()
plt.show()
```

```python
A_tensor = tf.constant(A, dtype=tf.float32)
S_tensor = tf.constant(S, dtype=tf.float32)
del A
```

```python
print(type(A_tensor))
print(A_tensor.dtype)
print(A_tensor.shape)

print(type(S_tensor))
print(S_tensor.dtype)
print(S_tensor.shape)
```

```python

class BMCallback(Callback):
	def __init__(self):
		self.best_weights = None
		self.min_rrmse = 100

	def on_epoch_end(self, epoch, logs=None):
		if logs['rrmse_gr'] < self.min_rrmse:
			self.min_loss = logs['rrmse_gr']
			self.best_weights = self.model.get_weights()

def loss_func(y_true, y_pred):
	y_pred = K.reshape(y_pred, (N**2, 1))
	S_pred = K.dot(A_tensor, y_pred)
	return K.mean(K.square(S_pred - S_tensor))


# def rrmse_gr(gr):
# 	def loss_func(y_true, y_pred):
# 		return K.sqrt(K.sum(K.square(y_pred - gr))/K.sum(K.square(gr)))
# 	return loss_func
```

```python
lrs = [0.001]
epochs = [500]
C = 1
gr_imgp = gr_img.reshape((N, N, C))
F = 16
depth = 4

np.random.seed(2)
# z = np.random.uniform(low=-1, high=1, size=(1,M,N,F))
z = np.zeros((1, N, N, F), dtype='float32')
bmcb = BMCallback()

# plt.figure(1)
# plt.figure(2)
print('started')
for i in range(1):
	for j in range(1):
		model = get_Model(Input((N, N, F)), n_filters=8, depth=depth, output_channels=C)
		model.compile(optimizer=Adam(lr=lrs[i], beta_1=0.90, beta_2=0.99), loss=loss_func)
		hist = model.fit(z, gr_imgp.reshape((1,N,N,C)), epochs=epochs[j], verbose=1)
		# plt.figure(1)
		# plt.subplot(3,3,3*i+j+1)
		# plt.imshow(rec_img[:,:,0], cmap='gray')
# 		model.set_weights(bmcb.best_weights)
		rec_img = model.predict(z)[0]
		print(f'lr_{lrs[i]}_ep_{epochs[j]}_loss_', hist.history['loss'][-1])

		print(f'rrmse(lr={lrs[i]}, epochs={epochs[j]}): {rrmse(gr_imgp, rec_img)}')
		# plt.figure(2)
		# plt.imshow(rec_img[:,:,0], cmap='gray')
		plt.imsave(f'logs/reconstruction/ChestPhantom{lrs[i]}_ep_{epochs[j]}.png', rec_img[:,:,0], cmap='gray')

# 		plt.plot(hist.history['rrmse_gr'])
# 		plt.show()
		# plt.savefig(f'logs/mri/mri_high_lr_{lrs[i]}_ep_{epochs[j]}_beta1_92_beta2_99.png')


#
# plt.title('Noisy')
# plt.imshow(noisy_img)
#
# plt.subplot(3,1,2)
# plt.title('Rec img')
# plt.imshow(j[0])
#
# plt.subplot(3,1,3)
# plt.title('Ground img')
# plt.imshow(gr_img)
# plt.figure(1)
# plt.savefig(f"logs/mri/mri_high1.png")
print('time: ', time.time()-t0)
# plt.show()
```
